```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
print(df.head())
```

```python
df.describe()
```

```python
# to include all categorial columns in the info
df.describe(include ='all')
```

```python
df.info()
```

```python
# to find the missing values in dataset
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df['age'].describe()
```

```python
df.columns
```

```python
#create a column for age_group
labels = [ 'Young Adult', 'Adult', 'Middle Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head()
```

```python
df[['customer_id','age','age_group','category','item_purchased','purchase_amount_(usd)']].head(10)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# to count number of purchases by each age group
age_purchase_count = df['age_group'].value_counts().sort_index()
print(age_purchase_count)
```

```python
#visualize the purchase counts by age group
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.barplot(x=age_purchase_count.index, y=age_purchase_count.values, palette='viridis')
plt.title('Number of Purchases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Purchases')
plt.show()
# to calculate average purchase amount by age group
```

```python
#create column purchase_frequency_days
frequency_mapping = { 'Fortnightly':14,
                     'Weekly':7,
                     'Monthly': 30, 'Quarterly': 90,'Bi-Weekly':14, 'Annually': 365,'Every 3 Months': 90
                     }
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
